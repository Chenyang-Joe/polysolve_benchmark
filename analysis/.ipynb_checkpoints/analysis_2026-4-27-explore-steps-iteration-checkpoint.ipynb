# Step / iteration exploration — 2026-04-27

Data: two **independent repeats** of a 60-step simulation on the 9996-row ball_rollers
mesh, with all four solvers running each newton iteration:

- `9996-full-run-test-repeat0`
- `9996-full-run-test-repeat1`

Same plot style as the last two cells of `analysis_2026-4-22_ball_rollers.ipynb`:
x-axis is `step-iter`, vertical dotted lines mark simulation-step boundaries.
The two repeats are drawn together so we can see (a) the per-solver pattern across
steps and (b) how reproducible the timings are across runs.

```python
import os, sys, re, contextlib, io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import importlib

sys.path.insert(0, '/u/1/chenyang/benchmark/analysis')
import libs.parser; importlib.reload(libs.parser)
from libs.parser import parse_log_file

SOLVER_COLORS = {
    'AMGCL':              '#d62728',
    'Eigen::PardisoLDLT': '#1f77b4',
    'Hypre_mpi':          '#2ca02c',
    'Trilinos':           '#ff7f0e',
}
ITERATIVE = ['AMGCL', 'Hypre_mpi', 'Trilinos']  # have num_iterations
SOLVERS   = list(SOLVER_COLORS.keys())

BASE = '/u/1/chenyang/benchmark_data/larger_matrix_exp/larger_mat_exp_result/2026-4-22'
REPEATS = ['9996-full-run-test-repeat0', '9996-full-run-test-repeat1']
DATA_SOURCE_NOTE = 'mesh: ball_rollers/9996  ·  60-step PolyFEM simulation, two independent repeats'


def _load_one(log_path, solver, repeat_tag):
    """Parse one log; suppress the mat_sz/nnz warnings (the source bin files were cleaned up)."""
    with contextlib.redirect_stdout(io.StringIO()), contextlib.redirect_stderr(io.StringIO()):
        entries = parse_log_file(log_path, solver)
    rows = []
    for e in entries:
        m = re.search(r'/(\d+)_(\d+)_A\.bin$', e.get('bin_A') or '')
        if not m:
            continue
        rows.append({
            'step':           int(m.group(1)),
            'iter':           int(m.group(2)),
            'elapse_time':    e.get('elapse_time'),
            'num_iterations': e.get('num_iterations'),
            'solver':         solver,
            'repeat':         repeat_tag,
        })
    return pd.DataFrame(rows)


frames = []
for rep in REPEATS:
    for s in SOLVERS:
        lp = os.path.join(BASE, rep, 'ball_rollers_8', f'{s}_9996.log')
        if not os.path.exists(lp):
            print(f'missing: {lp}'); continue
        df = _load_one(lp, s, rep)
        frames.append(df)
        print(f'  {rep:32}  {s:20}  rows={len(df):5}  steps={df.step.min()}-{df.step.max()}')

raw = pd.concat(frames, ignore_index=True)
print(f'\nTotal rows loaded: {len(raw)}')

# Shared x-axis: integer index over the union of (step, iter) pairs across all solver/repeat combos
unique_keys = (raw[['step', 'iter']]
               .drop_duplicates()
               .sort_values(['step', 'iter'])
               .reset_index(drop=True))
unique_keys['x'] = range(len(unique_keys))
key_to_x = {(s, i): x for s, i, x in zip(unique_keys['step'], unique_keys['iter'], unique_keys['x'])}
raw['x'] = [key_to_x[(s, i)] for s, i in zip(raw['step'], raw['iter'])]
step_start = unique_keys.groupby('step')['x'].min().to_dict()
print(f'Shared x-axis: {len(unique_keys)} unique (step, iter) pairs, steps {unique_keys.step.min()}-{unique_keys.step.max()}')
```

```python
def _decorate_step_axis(ax, df, ymax, label_step=True):
    """Draw step-boundary dividers + 'step N' annotations + tick labels."""
    for s in sorted(step_start.keys())[1:]:
        ax.axvline(step_start[s] - 0.5, color='grey', linestyle=':', linewidth=0.6, alpha=0.5)
    if label_step:
        for s, x0 in step_start.items():
            ax.text(x0, ymax * 1.04, f'step {s}', fontsize=6, rotation=90,
                    ha='left', va='bottom', color='grey')
    stride = max(1, len(unique_keys) // 60)
    tick_pos = list(unique_keys['x'][::stride])
    tick_lbl = [f"{unique_keys['step'].iat[i]}-{unique_keys['iter'].iat[i]}"
                for i in range(0, len(unique_keys), stride)]
    ax.set_xticks(tick_pos)
    ax.set_xticklabels(tick_lbl, rotation=90, fontsize=6)


def plot_y_vs_step_iter(df, y_col, solvers, title):
    """Scatter y_col vs shared x for each (solver, repeat) — repeat0 filled, repeat1 hollow."""
    fig, ax = plt.subplots(figsize=(18, 6))
    for solver in solvers:
        color = SOLVER_COLORS[solver]
        for repeat, marker, fill_alpha in [
            (REPEATS[0], 'o', 0.55),  # filled
            (REPEATS[1], 'x', 0.45),  # x-marker
        ]:
            g = df[(df['solver'] == solver) & (df['repeat'] == repeat)]
            if g.empty:
                continue
            tag = 'repeat0' if repeat == REPEATS[0] else 'repeat1'
            ax.scatter(g['x'], g[y_col], s=8, color=color, marker=marker, alpha=fill_alpha,
                       label=f'{solver} · {tag} (n={len(g)})')

    ymax = df[y_col].max()
    _decorate_step_axis(ax, df, ymax)
    ax.set_xlabel('step-iter (outer simulation step – inner newton iteration)')
    ax.set_ylabel(y_col)
    ax.set_title(f'{title}\n{DATA_SOURCE_NOTE}  ·  filled circles = repeat0, x = repeat1')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=8, loc='upper left', ncol=2)
    plt.tight_layout()
    return fig, ax
```

## Elapse time vs step-iter — all four solvers, both repeats

```python
fig, ax = plot_y_vs_step_iter(
    raw, 'elapse_time', SOLVERS,
    'Elapse time across simulation steps — 9996 mesh, OMP=8 / mpirun -np 8',
)
plt.show()
```

## `num_iterations` vs step-iter — iterative solvers only

`Eigen::PardisoLDLT` is direct so it has no iteration count; `AMGCL`, `Hypre_mpi`,
and `Trilinos` all use iterative solvers and report PCG / BiCGStab / similar
iteration counts.

```python
fig, ax = plot_y_vs_step_iter(
    raw, 'num_iterations', ITERATIVE,
    'Iterative-solver iteration count across simulation steps',
)
plt.show()
```
